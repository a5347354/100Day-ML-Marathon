```python
# 計算欄位的平均數 Mean
print("Mean value: %f" %app_train['AMT_INCOME_TOTAL'].mean())
#計算欄位的中位數 Median value
print("Median value: %f" %app_train['AMT_INCOME_TOTAL'].median())
#計算標準差 Sample standard deviation 
print("Sample standard deviation: %f" %app_train['AMT_INCOME_TOTAL'].std())
print(app_train['AMT_INCOME_TOTAL'].max())
```

```python
#畫直方圖
#直方圖：當設定 bins =10，會自動幫你把資料數值切成 10 等分，然後統計每一個數值內的資料數量
#年金/總收入
app_mean = app_train['AMT_ANNUITY']/app_train['AMT_INCOME_TOTAL']
app_mean.hist(bins=20)
```
